```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tag import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample document
document = "This is a sample document. It contains multiple sentences."

# Tokenization
tokens = word_tokenize(document)

# POS Tagging
pos_tags = pos_tag(tokens)

# Stopwords removal
stop_words = set(stopwords.words("english"))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]

# Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

# Join the preprocessed tokens into a string
preprocessed_text = " ".join(lemmatized_tokens)

# Calculate TF-IDF representation
corpus = [preprocessed_text]  # List of documents (in this case, only one document)
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)

# Get the feature names (terms)
feature_names = vectorizer.get_feature_names_out()

# Print the preprocessed text and TF-IDF representation
print("Preprocessed Text:")
print(preprocessed_text)
print("\nTF-IDF Representation:")
for i, feature_index in enumerate(tfidf_matrix.nonzero()[1]):
    term = feature_names[feature_index]
    tfidf_value = tfidf_matrix[0, feature_index]
    print("Term: '{}', TF-IDF: {:.4f}".format(term, tfidf_value))


from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer

s = "This is a simple sentence that I am using as an example in my Practical Examination. My name is Anurag."

word_token = nltk.word_tokenize(s)

print(word_token)

sent_token = nltk.sent_tokenize(s)

print(sent_token)

pos = nltk.pos_tag(word_token) # part-of-speech

pos

stop_word = set(stopwords.words("english"))

print(stop_words)

stopword_removed = [word for word in word_token if word.lower() not in stop_word]

stopword_removed

stemmer = PorterStemmer() # porter stemmer instance
stem_words = []
for i in stopword_removed:
    stem_words.append(stemmer.stem(i))
    

stem_words

lemmatizer = WordNetLemmatizer() #inctance of WordNetLemmatizer
lem_words = [lemmatizer.lemmatize(token) for token in stopword_removed]

lem_words

preprocessed_text = " ".join(lem_words)

preprocessed_text

corpus = [preprocessed_text]

vectorizer = TfidfVectorizer()
vecotr_matrix = vectorizer.fit_transform(corpus)

print(vecotr_matrix)

```
