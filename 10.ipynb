```python
'''### 10) Data Visualization III
Download the Iris flower dataset or any other dataset into a DataFrame. (e.g., https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the inference as:

1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a boxplot for each feature in the dataset.
4. Compare distributions and identify outliers.'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("iris.csv")
df.head()

df.shape

df["variety"].unique()

### 1. List down the features and their types (e.g., numeric, nominal) available in the dataset.

df.info()

### 2. Create a histogram for each feature in the dataset to illustrate the feature distributions.

sns.pairplot(df, diag_kind='hist',hue='variety' )
plt.show()

import seaborn as sns


sns.histplot(df['sepal.length'], kde=True)
sns.set_style('whitegrid')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.title('Histogram of Sepal Length')

plt.show()

sns.histplot(df['sepal.width'], kde=True)
sns.set_style('whitegrid')
plt.xlabel('Sepal Width')
plt.ylabel('Count')
plt.title('Histogram of Sepal Width')

plt.show()

sns.histplot(df['petal.length'], kde=True)
sns.set_style('whitegrid')
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.title('Histogram of Petal Length')

plt.show()

sns.histplot(df['petal.width'], kde=True)
sns.set_style('whitegrid')
plt.xlabel('Petal Width')
plt.ylabel('Count')
plt.title('Histogram of Petal Width')

plt.show()


import matplotlib.pyplot as plt

plt.hist(df['sepal.length'],density=True, bins=10)
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.title('Histogram of Sepal Length')

plt.show()

plt.hist(df['sepal.width'],density=True, bins=10)
plt.xlabel('Sepal Width')
plt.ylabel('Count')
plt.title('Histogram of Sepal Width')

plt.show()

plt.hist(df['petal.length'], density=True, bins=10)
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.title('Histogram of Petal Length')

plt.show()

plt.hist(df['petal.width'], density=True, bins=10)
plt.xlabel('Petal Width')
plt.ylabel('Count')
plt.title('Histogram of Petal Width')

plt.show()


### 3. Create a Boxplot for each feature in the dataset

sns.boxplot(x=df['sepal.length'])

import seaborn as sns

iris = sns.load_dataset('iris')

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
sns.boxplot(x=df['sepal.length'])
plt.xlabel('Feature')
plt.ylabel('Value')
plt.title('Boxplot of Sepal Length')

plt.subplot(2, 2, 2)
sns.boxplot(x=df['sepal.width'])
plt.xlabel('Feature')
plt.ylabel('Value')
plt.title('Boxplot of Sepal Width')

plt.subplot(2, 2, 3)
sns.boxplot(x=df['petal.length'])
plt.xlabel('Feature')
plt.ylabel('Value')
plt.title('Boxplot of Petal Length')

plt.subplot(2, 2, 4)
sns.boxplot(x=df['petal.width'])
plt.xlabel('Feature')
plt.ylabel('Value')
plt.title('Boxplot of Petal Width')

plt.tight_layout()
plt.show()


import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Sepal Length
sns.boxplot(x=df['sepal.length'], ax=axes[0, 0])
axes[0, 0].set_xlabel('Feature')
axes[0, 0].set_ylabel('Value')
axes[0, 0].set_title('Boxplot of Sepal Length')

# Sepal Width
sns.boxplot(x=df['sepal.width'], ax=axes[0, 1])
axes[0, 1].set_xlabel('Feature')
axes[0, 1].set_ylabel('Value')
axes[0, 1].set_title('Boxplot of Sepal Width')

# Petal Length
sns.boxplot(x=df['petal.length'], ax=axes[1, 0])
axes[1, 0].set_xlabel('Feature')
axes[1, 0].set_ylabel('Value')
axes[1, 0].set_title('Boxplot of Petal Length')

# Petal Width
sns.boxplot(x=df['petal.width'], ax=axes[1, 1])
axes[1, 1].set_xlabel('Feature')
axes[1, 1].set_ylabel('Value')
axes[1, 1].set_title('Boxplot of Petal Width')

# Adjust spacing between subplots
plt.tight_layout()

# Identify outliers
outliers = []
for feature in df.columns[:-1]:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    feature_outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    outliers.append(feature_outliers)

for idx, feature_outliers in enumerate(outliers):
    if not feature_outliers.empty:
        print(f"Outliers in {df.columns[idx]}:")
        print(feature_outliers)

# Show the plots
plt.show()


```
