```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

df.head()

df.shape

df.info()

df.describe()

df.corr()

plt.figure(figsize = (10,10))
sns.heatmap(
    data=df.corr(),
    cmap="coolwarm",
    center=None,
    robust=False,
    annot=True,
    fmt='.2g',
    annot_kws=None,
    linewidths=0,
    linecolor='white',
    cbar=True,
    cbar_kws=None,
    cbar_ax=None,
    square=False,
    xticklabels='auto',
    yticklabels='auto',
    mask=None,
)

sns.countplot(
    x='survived',
    hue='sex',
    data=df,
    saturation=0.75,
    dodge=True,
)
plt.title('Survival Count by Gender')
plt.show()

sns.countplot(
    x='survived',
    hue='class',
    data=df,
    saturation=0.75,
    dodge=True,
)
plt.title('Survival Count by class')
plt.show()

sns.histplot(data=df, x='fare', kde=True)
plt.title('Distribution of Ticket Fares')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()


sns.barplot(y='age', x='class',data=df)

sns.boxplot(
    x='class',
    y='fare',
    hue='sex',
    data=df,
    order=None,
    hue_order=None,
    orient=None,
    color=None,
    palette=None,
    saturation=0.75,
    width=0.8,
    dodge=True,
    fliersize=5,
    linewidth=None,
    whis=1.5,
    ax=None,
)

sns.histplot(data=df, x='fare')
plt.title('Distribution of Ticket Fares')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

sns.scatterplot(
    x='age',
    y='fare',
    hue='sex',
    style=None,
    size=None,
    data=df,
    palette=None,
    hue_order=None,
    hue_norm=None,
    sizes=None,
    size_order=None,
    size_norm=None,
    markers=True,
    style_order=None,
    x_bins=None,
    y_bins=None,
    units=None,
    estimator=None,
    ci=95,
    n_boot=1000,
    alpha=None,
    x_jitter=None,
    y_jitter=None,
    legend='auto',
)

dataset=['survived','pclass','age','fare','parch']
new_data = df[dataset].copy()

sns.pairplot(
    data=new_data,
    hue=None,
    hue_order=None,
    palette=None,
    vars=None,
    x_vars=None,
    y_vars=None,
    kind='scatter',
    diag_kind='auto',
    markers=None,
    height=2.5,
    aspect=1,
    corner=False,
    dropna=False,
    plot_kws=None,
    diag_kws=None,
    grid_kws=None,
    size=None,
)

```
