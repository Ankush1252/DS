```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

df.head()

df.isnull().sum()

df.info()

df.describe()

df.drop('deck', axis=1, inplace=True)

df['embark_town'].value_counts()

sns.countplot(x='embarked',data=df)

df['embark_town'].fillna('Southampton', inplace=True)

df['embarked'].fillna('S', inplace=True)

df['age'].fillna(df['age'].mean(), inplace=True)

df.isnull().sum()

df.head()

df.info()

df.head()

dummies = pd.get_dummies(df.sex)

df1 = pd.concat([df,dummies], axis=1)

df1.drop(['sex','male'], axis=1, inplace=True)

df1.head()

df1.drop('embark_town',axis=1, inplace=True)

df1.head()

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()


df1['label_embarked'] = labelencoder.fit_transform(df1.embarked)

df1['label_embarked'].value_counts()

df1.drop('embarked', axis=1, inplace=True)

df1.head()

df1.info()

```
